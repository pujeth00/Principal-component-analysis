```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from google.colab import drive 
drive.mount('/content/drive')
```

```python
path='/content/drive/My Drive/'
rd=pd.read_csv(path+'day2.csv')
df=pd.DataFrame(rd)
df
```

```python
X=np.array(df)
print(X)
```

```python
ean=np.mean(X,0)
print(ean)
```

```python
C=np.cov(X.T)
print(C)
```

Question 1

```python
from numpy import linalg as la 
w,v=la.eig(C)
print(w)
print(v)
```

```python
w.sort()
w
```

Question 2

```python
A=np.array(v[:4,:3])
B=np.array(v[:4,:2])
A=A.T 
B=B.T
print(A)
print(B)
```

```python
y=np.matmul(A,(X-ean).T)
print(y)
y1=np.matmul(B,(X-ean).T)
print(y1)
```

```python
xhat=np.matmul(A.T,y).T+ean 
print(xhat)
x1hat=np.matmul(B.T,y1).T+ean 
print(x1hat)
```

```python
df1=pd.DataFrame(xhat)
df1
```
